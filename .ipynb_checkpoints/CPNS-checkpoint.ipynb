```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = requests.get('https://sscn.bkn.go.id/public/data_noexist/pendaftar_dashboard_prod.js?_=1537378604051', verify = False)
```

```python
df = pd.DataFrame(a.json().get('data'))
```

```python
df
```

```python
cels = df.groupby('PENDIDIKAN_NM').nunique()
```

```python
def Sosiologi(data):
    if "SOSIOLOGI" in data:
        return "TRUE"
    else:
        return np.NaN
```

```python
df['YN'] = df.apply(lambda x: Sosiologi(x['PENDIDIKAN_NM']), axis=1)
```

```python
df.dropna(inplace=True)
```
